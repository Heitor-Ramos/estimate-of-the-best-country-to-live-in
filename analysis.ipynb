```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('World Happiness Report 2023.csv')
df.head()
```

```python
df = df.drop(columns=['upperwhisker', 'lowerwhisker'])

df.head()
```

```python
numeric_df = df.select_dtypes(include=[np.number])

mean_values = numeric_df.mean()
print("Médias:\n", mean_values)
```

```python
median_values = numeric_df.median()
print("Medianas:\n", median_values)
```

```python
std_values = numeric_df.std()
print("Desvios Padrão:\n", std_values)
```

```python
percentiles = numeric_df.quantile([0.25, 0.5, 0.75])
print("Percentis:\n", percentiles)
```
